```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete_df
```

```python
#total number of schools
unique_schools = len(school_data_complete_df["school_name"].unique())

unique_schools
```

```python
#total number of students
count_of_students = school_data_complete_df.count()[0]

count_of_students
```

```python
#total of budget
unique_budgets = school_data_complete_df['budget'].unique()

total_budget = unique_budgets.sum()

total_budget

#need to make into dollars
```

```python
#average math score

average_math_score = school_data_complete_df["math_score"].mean()

average_math_score
```

```python
#average reading score

average_reading_score = school_data_complete_df["reading_score"].mean()

average_reading_score
```

```python
#percentage of students with a passing math score (70 or greater) 
passing_math_score = len(school_data_complete_df.loc[school_data_complete_df["math_score"] >= 70])

percentage_passing_math_score = (passing_math_score/count_of_students)*100

percentage_passing_math_score
```

```python
#percentage of students with a passing reading score (70 or greater)
passing_reading_score = len(school_data_complete_df.loc[school_data_complete_df["reading_score"] >= 70])

percentage_passing_reading_score = (passing_reading_score/count_of_students)*100

percentage_passing_reading_score

```

```python
#% Overall Passing The percentage of students that passed math and reading.

total_passing_score = len(school_data_complete_df.loc[(school_data_complete_df["reading_score"] >= 70) &
                                                  (school_data_complete_df["math_score"] >= 70), :])


percentage_passing_score = (total_passing_score/count_of_students)*100

percentage_passing_score
```

```python
#Create a dataframe to hold the above results

district_summary_df = pd.DataFrame({"Total Schools": [unique_schools],
                          "Total Students": [count_of_students],
                          "Total Budget": [total_budget],
                          "Average Math Score": [average_math_score],
                          "Average Reading Score": [average_reading_score],
                          "% Passing Math": [percentage_passing_math_score],
                          "% Passing Reading": [percentage_passing_reading_score],
                          "% Overall Passing": [percentage_passing_score]})

district_summary_df
```

```python
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

district_summary_df
```

```python
#total students

total_students = school_data_complete_df["school_name"].value_counts()

total_students
```

```python
#total budget
total_budget = school_data_complete_df.groupby("school_name").mean()["budget"]

total_budget
```

```python
#average math scores

math_summary = school_data_complete_df.groupby("school_name").mean()["math_score"]

math_summary
```

```python
#average reading scores
reading_summary = school_data_complete_df.groupby("school_name").mean()["reading_score"]

reading_summary
```

```python
#school type 

school_types = school_data.set_index(["school_name"])["type"]

school_types
```

```python
#per student budget
per_school_budget = school_data_complete_df.groupby(["school_name"]).mean()["budget"]


per_student_budget = per_school_budget / total_students

per_student_budget
```

```python
#% passing reading

district_reading_score = school_data_complete_df.loc[school_data_complete_df["reading_score"] >= 70]

percentage_passing_reading = district_reading_score.groupby(["school_name"])["student_name"].count() / total_students * 100

percentage_passing_reading
```

```python
#% passing math

district_math_score = school_data_complete_df.loc[school_data_complete_df["math_score"] >= 70]

percentage_passing_math = district_math_score.groupby(["school_name"])["student_name"].count() / total_students * 100

percentage_passing_math
```

```python
#% overall passing
district_passing_score = school_data_complete_df.loc[(school_data_complete_df["reading_score"] >= 70) &
                                                  (school_data_complete_df["math_score"] >= 70), :]


district_passing_score = district_passing_score.groupby(["school_name"])["student_name"].count() / total_students * 100

district_passing_score
```

```python
summary_df = pd.DataFrame({"School Type": school_types,
                              "Total Students": total_students,
                              "Total School Budget": total_budget,
                              "Per Student Budget": per_student_budget,
                              "Average Math Score": math_summary,
                              "Average Reading Score": reading_summary,
                              "% Passing Math": percentage_passing_math,
                              "% Passing Reading": percentage_passing_reading,
                              "% Overall Passing": district_passing_score})
summary_df

```

```python
# top performing schools

top_schools = summary_df.sort_values(["% Overall Passing"], ascending=False)

top_schools.head(5)
```

```python
#bottom performing schools

bottom_schools = summary_df.sort_values(["% Overall Passing"], ascending=True)

bottom_schools.head(5)
```

```python
#by grade dataframe math

#set for each grade
ninth_graders = school_data_complete_df[(school_data_complete_df["grade"] == "9th")]
tenth_graders = school_data_complete_df[(school_data_complete_df["grade"] == "10th")]
eleventh_graders = school_data_complete_df[(school_data_complete_df["grade"] == "11th")]
twelfth_graders = school_data_complete_df[(school_data_complete_df["grade"] == "12th")]

#group by mean
ninth_graders_average = ninth_graders.groupby(["school_name"])["math_score"].mean()
tenth_graders_average = tenth_graders.groupby(["school_name"])["math_score"].mean()
eleventh_graders_average = eleventh_graders.groupby(["school_name"])["math_score"].mean()
twelfth_graders_average = twelfth_graders.groupby(["school_name"])["math_score"].mean()
```

```python
#math scores by grade

math_scores_summary_df = pd.DataFrame({"9th": ninth_graders_average,
                              "10th": tenth_graders_average,
                              "11th": eleventh_graders_average,
                              "12th": twelfth_graders_average})
math_scores_summary_df
```

```python
#by grade dataframe reading

#set for each grade
ninth_graders = school_data_complete_df[(school_data_complete_df["grade"] == "9th")]
tenth_graders = school_data_complete_df[(school_data_complete_df["grade"] == "10th")]
eleventh_graders = school_data_complete_df[(school_data_complete_df["grade"] == "11th")]
twelfth_graders = school_data_complete_df[(school_data_complete_df["grade"] == "12th")]

#group by mean
ninth_graders_average = ninth_graders.groupby(["school_name"])["reading_score"].mean()
tenth_graders_average = tenth_graders.groupby(["school_name"])["reading_score"].mean()
eleventh_graders_average = eleventh_graders.groupby(["school_name"])["reading_score"].mean()
twelfth_graders_average = twelfth_graders.groupby(["school_name"])["reading_score"].mean()
```

```python
#reading scores by grade

reading_scores_summary_df = pd.DataFrame({"9th": ninth_graders_average,
                              "10th": tenth_graders_average,
                              "11th": eleventh_graders_average,
                              "12th": twelfth_graders_average})
reading_scores_summary_df
```

```python
#scores by schools spending

bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

summary_df["Spending Ranges (Per Student)"] = pd.cut(x = summary_df["Per Student Budget"], bins = bins, labels = labels)

summary_df
```

```python
#create new data frame for scores by spending

scores_by_spending_df = summary_df.groupby("Spending Ranges (Per Student)")

print(scores_by_spending_df.mean())

scores_by_spending_df[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]].mean()
```

```python
#scores by school size

bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

summary_df["School Size"] = pd.cut(x = summary_df["Total Students"], bins = bins, labels = labels)

summary_df
```

```python
#create new data frame for school size

scores_by_size_df = summary_df.groupby("School Size")

print(scores_by_size_df.mean())

scores_by_size_df[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]].mean()
```

```python
#scores by school type

scores_by_type_df = summary_df.groupby("School Type")

print(scores_by_type_df.mean())

scores_by_type_df[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]].mean()
```
